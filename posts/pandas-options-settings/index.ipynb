```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(10, 10), columns=list('ABCDEFGHIJ')) * 10**7
df
```

```python
# supress scientific notation and control decimal points shown with float_format
pd.options.display.float_format  = '{:.2f}'.format
df
```

```python
# control max rows and columns
pd.options.display.max_rows = 4
pd.options.display.max_columns = 4
df
```

```python
# large_repr lets you select whether to display dataframes that exceed max_columns or max_rows
# as a truncated frame ('truncate'), or as a summary ('info'). 'truncate' is default
pd.options.display.large_repr = 'info'
df
```

```python
# reset options
pd.reset_option('all', silent=True)
df.head()
```
