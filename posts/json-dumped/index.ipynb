```python
import os
from collections import OrderedDict
import pythoncom
from win32com.shell import shell, shellcon

def shortcut_target(filename):
    # from http://timgolden.me.uk/python/win32_how_do_i/read-a-shortcut.html
    link = pythoncom.CoCreateInstance(
        shell.CLSID_ShellLink,    
        None,
        pythoncom.CLSCTX_INPROC_SERVER,    
        shell.IID_IShellLink
    )
    link.QueryInterface(pythoncom.IID_IPersistFile).Load(filename)
    name, _ = link.GetPath(shell.SLGP_UNCPRIORITY)
    return name

```

```python
os.chdir(r"C:\Users\rhohe\AppData\Roaming\Microsoft\Windows\Recent")
shortcuts = [i for i in os.listdir() if i[-4:]=='.lnk']
shortcuts.sort(key=os.path.getctime, reverse=True)
folders = []
for shortcut in shortcuts:
    target = shortcut_target(shortcut)
    folders.append(os.path.split(target)[0])
folders = list(OrderedDict.fromkeys(folders).keys())
for i, folder in list(enumerate(folders))[::-1]:
    print(i, folder)
i = int(input())
os.chdir(folders[i])
os.system('start cmd')
```
